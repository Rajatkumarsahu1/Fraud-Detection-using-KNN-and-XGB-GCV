```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
import xgboost as xgb
```

```python
from sklearn.datasets import load_wine
```

```python
load_wine
```

```python
url='https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-white.csv'
df=pd.read_csv(url,sep=';')
```

```python
df.head()
```

```python
df['quality'].unique()
```

```python
df.info()
```

```python
df['quality']=df['quality'].apply(lambda x:1 if x>=7 else 0)
```

```python
df.head()
```

```python
df.columns
```

```python
x=df[['fixed acidity', 'volatile acidity', 'citric acid', 'residual sugar',
       'chlorides', 'free sulfur dioxide', 'total sulfur dioxide', 'density',
       'pH', 'sulphates', 'alcohol']]
y=df['quality']
```

```python
std=StandardScaler()
x_std=std.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_std,y,test_size=0.2,random_state=35)
```

```python
model_xgb=xgb.XGBClassifier(objective='binary:logistic',eval_metric='logloss')
```

```python
model_xgb.fit(x_train,y_train)
```

```python
model_xgb.predict(x_test)
```

```python
print("classification_report \n",classification_report(y_test,model_xgb.predict(x_test)))
```

```python
accuracy_score(y_test,model_xgb.predict(x_test))
```

```python
confusion_matrix(y_test,model_xgb.predict(x_test))
```

```python
param_grid = {
    'n_estimators': [50,100,150, 250,],
    'max_depth': [3,5,7],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'learning_rate':[0.01,0.1,0.15,0.2],
    "subsample":[0.7,0.8,1],
    'colsample_bytree':[0.7,0.8,0.1]
}
```

```python
grid_search_cv=GridSearchCV(model_xgb,param_grid,cv=3,scoring='accuracy',n_jobs=-1,verbose=1)
```

```python
grid_search_cv.fit(x_train,y_train)
```

```python
grid_search_cv.best_params_
```

```python
model_xgb_best_param=xgb.XGBClassifier(**grid_search_cv.best_params_,objective='binary:logistic',eval_metric='logloss')
```

```python
model_xgb_best_param.fit(x_train,y_train)
```

```python
accuracy_score(y_test,model_xgb_best_param.predict(x_test))
```

```python
print("classification_report \n",classification_report(y_test,model_xgb_best_param.predict(x_test)))
```

```python
confusion_matrix(y_test,model_xgb_best_param.predict(x_test))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.datasets import load_iris
```

```python
data=load_iris()
```

```python
data
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
df['target']=data.target
```

```python
df.head()
```

```python
df.columns
```

```python
x=df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)',
       'petal width (cm)']]
y=df['target']
```

```python
x_std=std.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_std,y,test_size=0.2,random_state=35)
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(x_train,y_train)
```

```python
knn.predict(x_test)
```

```python
accuracy_score(y_test,knn.predict(x_test))
```

```python
param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}
```

```python
knn = KNeighborsClassifier()

grid_search = GridSearchCV(estimator=knn,
                           param_grid=param_grid,
                           cv=5,
                           scoring='accuracy',
                           verbose=1,
                           n_jobs=-1)

grid_search.fit(x_train, y_train)
```

```python
print("✅ Best Parameters:", grid_search.best_params_)
print("🏆 Best CV Accuracy:", grid_search.best_score_)
```

```python
best_knn = grid_search.best_estimator_
y_pred = best_knn.predict(x_test)
print("🎯 Test Accuracy:", accuracy_score(y_test, y_pred))
```
