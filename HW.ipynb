```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
df=pd.read_csv('creditcard.csv')
```

```python
plt.figure(figsize=(12, 8))
corr = df.corr()
sns.heatmap(corr[['Class']].sort_values(by='Class', ascending=False), annot=True, cmap='coolwarm')
plt.title('Correlation of Features with Class')
plt.show()
```

```python
# Distribution of Amount for fraud vs legit
plt.figure(figsize=(8, 5))
sns.boxplot(x='Class', y='Amount', data=df)
plt.title('Transaction Amount by Class')
plt.show()
```

```python
# Time vs Class scatter (optional)
plt.figure(figsize=(10, 5))
sns.scatterplot(x='Time', y='Amount', hue='Class', data=df, alpha=0.5)
plt.title('Transaction Amount over Time by Class')
plt.show()
```

```python

fig, axes = plt.subplots(1, 2, figsize=(12, 5))
sns.histplot(df['Amount'], bins=50, ax=axes[0])
axes[0].set_title("Amount Distribution")
sns.histplot(df['Time'], bins=50, ax=axes[1])
axes[1].set_title("Time Distribution")
plt.tight_layout()
plt.show()
```

```python

X = df.drop('Class', axis=1)
y = df['Class']

scaler = StandardScaler()
X[['Amount', 'Time']] = scaler.fit_transform(X[['Amount', 'Time']])

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_params = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}
knn = KNeighborsClassifier()
knn_grid = GridSearchCV(knn, knn_params, cv=3, scoring='f1', verbose=1, n_jobs=-1)
knn_grid.fit(X_train, y_train)
print("Best KNN Params:", knn_grid.best_params_)

y_pred_knn = knn_grid.best_estimator_.predict(X_test)
print("KNN Classification Report:")
print(classification_report(y_test, y_pred_knn))
```

```python
import xgboost as xgb
xgb_params = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5],
    'learning_rate': [0.05, 0.1]
}
xgb = xgb.XGBClassifier(use_label_encoder=False, eval_metric='logloss')
xgb_grid = GridSearchCV(xgb, xgb_params, cv=3, scoring='f1', verbose=1, n_jobs=-1)
xgb_grid.fit(X_train, y_train)
print("Best XGBoost Params:", xgb_grid.best_params_)

y_pred_xgb = xgb_grid.best_estimator_.predict(X_test)
print("XGBoost Classification Report:")
print(classification_report(y_test, y_pred_xgb))
```
